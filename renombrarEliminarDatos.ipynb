```python
import pandas as pd
```

```python
filename = '/content/drive/My Drive/DataScience/BasesDeDatos/RailsToTrails_National Count Data_week.xlsx'
bike_df = pd.read_excel(filename)
bike_df.head()
```

## Eliminar las columnas innecesarias

```python
bike_df = bike_df.drop(columns = ['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'])
bike_df.head()
```

## Renombrar la columna Unnamed: 6 a percent_change

```python
bike_df = bike_df.rename(columns = {'Unnamed: 6': 'percent_change'})
bike_df.head()
```

## ¿Cuántos valores faltantes hay en el conjunto de datos? ¿En qué columnas están?

```python
bike_df.isna().sum()
```

Respuesta: Hay 52 datos faltantes en la columna 2021 Counts

## Rellenar los valores faltantes en la columna 2021 Counts con 0

```python
bike_df['2021 Counts'].fillna(0, inplace = True)
```

```python
bike_df['2021 Counts'] = bike_df['2021 Counts'].astype(int)
```

```python
bike_df.head()
```
