```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/MyDrive/DataScience/BasesDeDatos/kingCountyHouseData.csv'
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
df.loc[df['price'] <= 300000]
```

```python
price_filter = df.loc[:, 'price'] <= 3000000
price_filter
```

```python
df.loc[price_filter, 'price'].hist(bins = 30, edgecolor='black')
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
# Adding seaborn style 
plt.style.use('seaborn')
price_filter = df.loc[:, 'price'] <= 3000000

df.loc[price_filter, 'price'].hist(bins = 30, edgecolor='black')
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```
