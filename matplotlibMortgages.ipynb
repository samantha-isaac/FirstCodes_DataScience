```python
import matplotlib.pyplot as plt
import pandas as pd
```

## 2. Creen un filtro para una tasa de interés del 3 % (observen cómo está escrito en los datos).  

```python
df.loc[df['Interest Rate']  == .03]
```

## 3. Definan el nuevo DataFrame basándose en los filtros de año e interés que han creado.  

```python
df_30years_3InterestR = df.loc[(df['Interest Rate'] == .03) & (df['Mortgage Name'] == '30 Year')]
df_30years_3InterestR
```

## 4. Tenemos una opción para el 5 % de tasa de interés:

```python
# el formato del símbolo ~ para negar los resultados del filtro.
df_30years_5InterestR = df.loc[((df['Interest Rate'] == .05) & (df['Mortgage Name'] == '30 Year'))]
df_30years_5InterestR
```

## 5. Puesto que queremos graficar la suma acumulativa del interés pagado, podemos usar la función .cumsum().   ¿Cuáles son los tres primeros valores en la salida para la suma acumulativa de la tasa del 3 %?

```python
df_30years_5InterestR['Interest Paid'].cumsum().head(3)
```

## 6. Usen plt.plot(x, y, c=color) para trazar la línea negra (5 %).  Queremos los valores x para que sean ['Month'] y los valores y para que sean la suma acumulativa del Interest Paid (interés pagado).  Podemos usar el código de la pregunta anterior para definir los valores y dentro de la función plt.plot().  Pista: para el negro, c = 'k'

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10,5))
plt.plot((df_30years_3InterestR['Month']), (df_30years_3InterestR['Interest Paid'].cumsum()), c='k')
plt.plot((df_30years_5InterestR['Month']), (df_30years_5InterestR['Interest Paid'].cumsum()), c='b');
```
